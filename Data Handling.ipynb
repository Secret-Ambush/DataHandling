```python
np.ran
x1 = 100 * np.random.rand(100)
x2 = 50 * np.ones(50)
x3 = 100 * np.random.rand(100) + 100
x4 = -100 * np.random.rand(100)
data = np.concatenate((x1,x2,x3,x4))
```

```python
outlier = []
def detect_outliers(data):
  threshold = 3
  mean = np.mean(data)
  std = np.std(data)
  for i in data:
    z_score = (i-mean)/std
    if np.abs(z_score)>threshold:
      outlier.append(i)
  return outlier

outpts = detect_outliers(data)
outpts
```
